## 1. Introduction, background & research questions

### 1.1. Introduction

In 2021, 2846 deaths in England occurred as a result of poisoning from drug misuse, an 80% rise on 2011 levels (ONS, 2022). Looking at this figure, it would be reasonable to assume that the prevalence of drug misuse itself must also be rising. Yet, in 2021, the proportion of individuals reporting drug use in the previous year remained steadily below 10%, and represents the continuation of a relatively stable trend since 2011 (ONS, 2022). So, if drug misuse prevalence is falling, yet deaths from drug misuse are rising, what is behind this disparity? 

### 1.2. Background & literature review

It is well documented that drug misuse blights individuals, families, and communities across England, with wide ranging impacts that span social, political, economic and health domains. Recent analysis by Hughes et.al. (2019) estimates that Â£4.9 billion of government expenditure is annually attributable to drug misuse in England, whilst other research highlight impacts on mortality (Heazell et.al., 2020), domestic and sexual abuse (Berbow et.al., 2018), and modern slavery (Castro-Gonazalez, 2022). Moreover, large bodies of literature have identified that drug misuse is also a particularly pertinent urban challenge, one that exacerbates ethnic inequalities and threatens community cohesion (Flacks, 2018; Fatsis, 2019; Vomfell and Stewart, 2021), perpetuates deprivation (Newton et.al., 2015; Lower et.al., 2022), and drives urban crime (Spicer et.al., 2020). 

Comparatively less literature explores the deaths that arise from drug misuse. This is surprising given the record levels seen in recent years, and surprising because each death represents the tragic loss of a loved one or friend. Yet, importantly, each death also represents an urban challenge: the systemic failure of health, social, educational, and political systems and safety nets. To date, research has identified associations between drug misuse deaths and age (Kontopantelis et.al., 2018), gender (King et.al., 2014), ethnicity (Harrison et.al., 1997), social class and deprivation (Esmail et.al., 1993; Mcphee et.al., 2018; John et.al., 2019), and education (Fletcher and Hargreaves, 2008). Studies have also explored geographical variations in drug misuse deaths, with Griffiths et.al. (2008) finding that urban areas in England experience higher mortality rates, and Congdon (2020) investigating the geographical nature of drug misuse deaths in the USA.

Internationally, there is a much larger scope of work: studies in the USA have assessed the impact of employer interventions (Akanbi et.al., 2020), gangs (Johnson et.al., 2020), and citizenship issues (Pinedo, 2020) upon drug misuse deaths- albeit more often with a focus on fentanyl and opioids that are a greater concern in the USA, but which makes their findings less applicable to an English context. Though, most work takes a one-dimensional approach to drug misuse deaths, examining the impact of a single factor upon mortality, or examining the risk factors for drug misuse deaths from only one drug- this is largely true for the aforementioned literature. This approach arguably fails to capture the complex and multi-faceted nature of drug misuse deaths. Moreover, much literature is outdated and not reflective of current conditions in which drug misuse deaths are reaching record levels even as drug misuse prevalence declines. 

Without a contemporary and holistic understanding of the drivers and associations of drug misuse deaths, policy making, resource allocation and health and social care provision cannot be adequately targeted nor successful. Such targeted resource allocation and outreach services are especially pertinent in a time of constrained health and local government budgets (Harlock et.al., 2019), reductions in which have been shown to contribute to higher rates of drug misuse deaths (Mcphee and Sheridan, 2020). As such, this project seeks to contribute to this literature gap by providing a contemporary analysis of English drug misuse deaths in 2021.

## 2. Data sources

To gain a holistic and encompassing understanding of the causes behind drug misuse deaths, various demographic and socioeconomic factors must be analysed. In the case of factors such as age, sex, education and deprivation, these were selected based on the review and findings of existing literature. For factors such as fuel poverty and housing energy efficiency, these were chosen given the current economic and political climate in which these are all pertinent social challenges and may affect recent drug misuse deaths. 

There is, however, no single dataset that contains information pertaining to a drug misuse death along with demographic, socioeconomic, and contextual information for that death. Rather, the Office for National Statistics (ONS) publishes annual data of the number of drug misuse deaths at a national scale, broken down by age and gender. Alongside this, the ONS also publishes a breakdown of annual drug misuse deaths by local authority- though this is not stratified by age, gender, or any other demographic variable. This means that, at a local authority level, drug misuse deaths cannot be analysed based on gender or age. They can, however, be analysed with regard to the demographic and contextual characteristics of the local authority. Instead, drug misuse deaths as a function of age and gender must be analysed at the national level for which the data is available. As there is not one combined dataset, to gain an understanding of the impacts of demography and socioeconomics upon drug misuse deaths at a local authority level, various data must be sourced and joined. 

**Data sources:**

**1. Drug misuse deaths by age and gender-** this is an Excel file of the count of recorded drug misuse deaths for the whole of England between 1993 and 2021. Each row corresponds to one year. One sheet contains a column for total deaths (both genders), a column for male deaths, and a column for female deaths. Another sheet contains the count of recorded deaths stratified by age band (Under 20, 20-29, 30-39, and so on). The data is an annual release, compiled by the ONS and collected from Local Registration Services (LRS). Where a death occurs towards the end of a reporting year, and is registered late (due to coroner or police involvement), the death will be recorded in the next reporting year. Nonetheless, this is the most reliable, comprehensive and accurate data source of national drug misuse prevalence. It will contribute towards Research Question 1.

We must sum the counts of respondents in these categories in each local authority, and additionally sum the total number of respondents across all ethnicity categories in each local authority to give a total population from which a percentage can be derived.

```python
# Grouping all 19 ethnicities in an LA and aggregating the sum of all respondent counts, to get a 
# total respondent population for each LA
pop_eth = ethnicity_LA.groupby(['Lower Tier Local Authorities Code'], as_index=False).agg({'Observation':['sum']})
pop_eth.columns = ['Lower Tier Local Authorities Code','Sum_total_pop']
```

Given that a simple moving average time series produces an average value for each time point based on previous observations in a given time window, a 2 year period was chosen as it will reflect short term variability in the data. On the other hand, a 5 year period reflects more of the historic variation, whilst a cumulative average accounts for variation across the entire 10 year period (2011-2021). 

It is clear that the most variable trend is that of the 2 year average, which exhibits the largest peaks and troughs when compared to the 5 year and cumulative averages. This is expected, given that the two-year period means the trend is more susceptible to recent observations and so will likely follow a similar trajectory to the raw data. The shallow gradient of the 2 year average between 2016 and 2018, a period in which drug misuse deaths rapidly increased, suggests that the trend follows a slight lag behind the actual number of deaths. In contrast, the 5 year and cumulative average time series display much less variation because of the smoothing effect of a larger time period on individual years. Between 2011 and 2015, the two are similar, though the 5 year average increases after this point. Overall, the time series trends indicate that the annual number of drug misuse deaths have, on average, increased between 2011-2021.

To assess whether there are statistically significant differences in drug misuse deaths between age bands, we will use a Kruskal-Wallis test. This is a non-parametric statistical test that compares the medians of at least three independent samples. This test is used instead of a parametric ANOVA method because, as we saw in Section 3.5, 5/6 variables are normally distributed, but the age band '30-39' is not. Thus, a parametric ANOVA test would not be suitable in this instance.

A Kruskal-Wallis test only identifies if there are any significant differences between the groups as a whole. It does not indicate the groups that such differences occur between. So, a post-hoc Dunn test must also be performed to identify statistically significant differences between individual groups. 

```python
# Performing the Kruskal-Wallis test 
stats.kruskal(national_df['Under_20'], national_df['Twenty_29'], national_df['Thirty_39'],
                                        national_df['Forty_49'], national_df['Fifty_69'], national_df['Over_70'])
```

The Kruskal-Wallis test indicates that there are significant differences amongst the 6 age groups (Kruskal-Wallis, p<0.05).

```python
# For the post-hoc Dunn test, the data must be hashable- one way to ensure this is by converting dataframe
# columns to lists
Under_20 = national_df['Under_20'].to_list()
Tw_29 = national_df['Twenty_29'].to_list()
Th_39 = national_df['Thirty_39'].to_list()
Fo_49 = national_df['Forty_49'].to_list()
Fi_69 = national_df['Fifty_69'].to_list()
Over_70 = national_df['Over_70'].to_list()
```

```python
# These lists are then joined into one list of data
all_bands = [Under_20, Tw_29, Th_39, Fo_49, Fi_69, Over_70]
```

```python
# Performing the post-hoc Dunn test
sp.posthoc_dunn(all_bands, p_adjust = 'bonferroni')
```

From the Dunn post-hoc test, it is clear that there are statistically significant differences (Dunn, p<0,05) in drug misuse deaths between:
- Under 20's and 20-29, 30-39, 40-49, and 50-69 groups
- 20-29's and the 40-49 group 
- Over 70's and the 30-39, 40-49, and 50-69 groups

### 4.4. "_What are the temporal trends in drug misuse deaths? (RQ.4)_"

To understand whether there are temporal trends in drug misuse deaths across England, and to make future predictions, we will first use Single Exponential Smoothing to calculate the optimal smoothing level for the data and capture variation in the level component of the data. Finally, Double Exponential Smoothing will be performed to capture the variation in both the level and trend components of the data, and to make future predictions.

#### 4.4.1. Single exponential smoothing

```python
# Making the total number of deaths into a list
deaths_list = national_df['Deaths'].tolist() 
# Creating a date-time index
dates = pd.DatetimeIndex(national_df.Year)
# Joining the list to the date-time index as an array
deaths_ESA = pd.Series(deaths_list, dates)
# Checking that this has worked
display(deaths_ESA.head())
```

```python
# Calculating the optimal smoothing level
ESAfit1 = SimpleExpSmoothing(deaths_ESA, initialization_method="estimated").fit()
# Calculating predictions for the next 3 years based on this optimal smoothing level
fcast1 = ESAfit1.forecast(3).rename(r'$\alpha=%s$'%ESAfit1.model.params['smoothing_level'])
```

```python
# Creating the plot
plt.figure(figsize=(10, 4))
# Plotting the original total number of deaths in dashed, transparent black
raw_line_1, = plt.plot(deaths_ESA, color='black', alpha = 0.5, linestyle = 'dashed')
# Plotting the exponential smoothing averages model that uses the optimal smoothing level
line1 = plt.plot(ESAfit1.fittedvalues, color='green')
# Plotting the 3 year predictions of this model
line1, = plt.plot(fcast1, color='green')
# Adding an appropriate legend
plt.legend([line1, raw_line_1], [fcast1.name, 'Raw data'], edgecolor = 'white')
# Adding axis labels and title
plt.ylabel("Deaths", fontsize=10, labelpad=10)
plt.xlabel("Year", fontsize=10, labelpad=10)
plt.title("Exponential smoothing average of total drug misuse deaths using the optimal smoothing level", fontsize=10)
# Adding a lightly transparent grid, to aid readability
plt.grid(alpha = 0.2)
# Removing right and top axis spines
sns.despine()
```

Above, the optimal alpha parameter is calculated as 0.995. Alpha is a smoothing function that captures the level component of the data at each time point. Given that alpha values range from 0-1, this is a high alpha value. Higher alpha values depreciate faster over the previous time points than smaller values do, and so higher alpha values assign less weight to historical observations. This means that the average is likely to be more variable, because it assigns more weight to recent observations and so is more susceptible to existing variation in the data. 

Such variability is visible in the plot, where the trajectory of the average closely follows that of the raw data, exhibiting slightly lagged, but similar, peaks and troughs. This can be seen in the smoothing average in 2018, where there is a sharp decrease in drug misuse deaths that mirrors the drop in the raw data in 2017. This, again, is because the recent observations for each time point are weighted more strongly than historical observations, meaning recent observations will be more influential in the calculation of the smoothed average for each time point. As a result of this, the single exponential smoothing does not provide much insight or value, for it fails to capture much of the variation in the level component of the data. Thus, the predictive ability of the model is severely limited.

In the first plot, the trend is slightly smoother and less variable with fewer, and less sharp, peaks and troughs. It also visually follows the trend of the raw data to a greater extent than the other beta coefficients, particularly around 2013 and 2019. This can also be seen in the second figure, in which the 0.2 beta coefficient (the green line) displays the least fluctuations. This is because 0.2 is a low beta coefficient and so, for each time point, it weights the historical observations to a greater extent than a higher beta coefficient does, because lower values depreciate more slowly. This means that more historical observations, and thus trends and level components, are incorporated into the model when using a lower beta coefficient, meaning the average is more historically comprehensive and smooth, and that more trend variation is captured by the model. 

In terms of model predictions, a beta parameter of 0.2 produces the steepest prediction, followed by the parameters 0.5, 0.8, and 1.0 This is because the use of a low beta parameter means that the model captures more historical variation in the trend and so differs from those of higher beta values. As such, this prediction is arguably more representative of the data across the whole period than predictions using higher beta values. Such a low beta parameter is suitable for this analysis, in which we aim to make predictions based on the trends in drug misuse deaths over the entirety of the 2011-2021 period, because it captures more of the historical temporal variation and trends in the data. Thus, 0.2 is identified as the optimal beta parameter.

## 5. Discussion

**5.1.** _**"How have drug misuse deaths changed over time, and are there differences in drug misuse deaths between genders and between age groups?" (RQ.1)**_

Results indicate foremostly that drug misuse deaths increased by 80% between 2011-2021, with annual increases occurring in all years except two. This upwards trend is consistent with the 2020 work of the ONS, which found a similar rate of change between 2012-2020 (Breen and Manders, 2020), and sets a worrying tone for the future. It was also found that there are statistically significant differences in drug misuse deaths between males and females (t-test, p<0.05), with males experiencing almost 90% higher deaths over the time period. Such gender disparities in drug misuse deaths are well documented and support these findings: Congdon (2020) identified 70% higher drug misuse deaths amongst US males than females, whilst Parkinson et.al. (2018) found a similar disparity in Scotland. Tweed et.al. (2020) describes how the causes of this gender disparity are uncertain, though Stein et.al. (2020) states that unemployment is partly to blame with men more susceptible to economic restructuring that increases the likelihood of unemployment. 

Similarly, significant differences in drug misuse deaths between age bands were identified (Kruskal-Wallis, p<0.05). Further analysis revealed that these significant differences exist between those aged under 20 and those aged 20-29, 30-39, 40-49, and 50-69, between those aged 20-29 and those aged 40-49, and between those aged over 70 and those aged 30-39, 40-49, and 50-69 (Dunn, p<0.05). These differences are visually illustrated in the bar chart in Section 4.1.4, which shows how those aged 40-49 experience the highest number of drug misuse deaths between 2011-2021, while the line chart in the same section shows that those aged 30-39 also experience similarly high levels. In contrast, those aged under 20 and over 70 experienced far fewer drug misuse deaths. It has long been documented that middle aged individuals are the most susceptible to deaths from drug misuse (Bird et.al., 2004; Webb et.al., 2003; Middleton and Stringer, 2016), supporting these findings, but there is comparatively sparse literature that explains why this is the case. Middleton and Stringer (2016) suggest that physical frailty is behind the increased deaths seen amongst older age bands, though Morgan et.al. (2006) suggest that the differences are not cohort based, rather they are because of the size of the drug using population of each age band and that, naturally, this size is greater for those middle aged. Nonetheless, these results are supported by existing literature, and identify statistically significant differences in drug misuse deaths between age bands.

Additionally, it was identified that there is no statistically significant association between drug misuse deaths in England between 2011-2021 and drug misuse prevalence in England and Wales over the same time period (Linear Regression, p= 0.096). This is surprising in the context of rising drug misuse deaths: it would intuitively make sense that for drug misuse deaths to rise, drug misuse must also rise. As we can see in Section 4.1.5, however, this is not the case, rather the two trends follow visually unrelated trajectories. One reason for this may be due to the data comparability. Drug misuse prevalence data incorporated both England and Wales, whilst the number of drug misuse deaths is only applicable to England, meaning that the datasets were not entirely comparable. Though, in any case, the higher counts of drug misuse deaths with the inclusion of Welsh data should mean that the linear model would produce a stronger regression, so this is likely a minimal implication. Additionally, the drug misuse prevalence data does not account for individuals living in student accommodation, and so likely fails to capture a proportion of drug misuse which may potentially affect these results. 

**5.2.** "_**Do drug misuse deaths vary spatially across England?" (RQ.2)**_

Results also indicate that, in 2021, drug misuse deaths vary spatially amongst lower-tier LA in England. Spatial clustering and spatial lag models demonstrate that there are clusters in which LA with high drug misuse deaths neighbour LA with similarly high drug misuse deaths, and vice-versa. These are particularly prominent in the North and East, especially in urban areas such as Birmingham and Leeds, and this finding is supported by the statistically significant Moran's I value of 0.166 (p<0.05), which indicates a level of spatial autocorrelation. This clustering may be attributable to neighbouring LA sharing similar characteristics that promote drug misuse, meaning neighbouring LA experience similar levels of drug misuse deaths. Freisthler et.al. (2005) identified that drug use and drug sales are spatially autocorrelated over small area geographies, giving weight to this possibility. Though existing literature on the spatial autocorrelation of drug misuse is sparse, Lorant et.al. (2001) found that health events and outcomes, morbidity, and mortality are spatially autocorrelated across small area geographies in Belgium, suggesting that so too are other ecological factors that may drive drug misuse deaths. Thus, it is clear that there is spatial variation and clustering within drug misuse deaths in 2021 in England.

**5.3.** _"**How are socioeconomic and demographic variables associated with drug misuse deaths, and is there spatial variability in these associations?" (RQ.3)**_

To gain an understanding of the factors behind drug misuse deaths, the associations between drug misuse deaths in 2021 and demographic and socioeconomic factors were assessed using Linear Regression. The final model captured 71.3% of variation in drug misuse deaths (Adj R2 = 0.713; AIC = 1945), and highlights a number of notable associations. One is the statistically significant association between income deprivation score and drug misuse deaths (p<0.05), in which a one unit increase results in an increase of 49.1 drug misuse deaths. This finding is supported by the work of Alexiou et.al. (2021) and Wilkinson and Pickett (2007), though it is in contrast to Graham et.al. (2012) who found that drug related mortality is not affected by deprivation. The reason for this contrast may be that this project analysed recent data (2011-2021) whereas Graham et.al. (2012) looked at a period pre-2010. Additionally, Graham et.al. (2012) focused on Belfast and Glasgow, two cities with different challenges, and did not consider the entirety of England. 

Another notable association is that of JSA Claimant Count and drug misuse deaths. JSA Claimant Count was used as a proxy for unemployment and is significantly associated with drug misuse deaths (p<0.05). Though, the strength of the association is weak- a one unit increase in JSA claimant count in an LA results in an increase of 0.0013 in drug misuse deaths in 2021. This contrasts to Griffiths et.al. (2008) who found that drug misuse deaths are significantly more common amongst the unemployed. This disparity may be due to the use of different metrics or the differing time periods. Overall, the results of the linear model identified significant associations between drug misuse deaths and income deprivation score, JSA claimant count, the proportion of residents of white ethnicity, the proportion of households owned with a mortgage, the proportion of residents with a degree, and the proportion of households privately rented.

As it was ascertained that drug misuse deaths were spatially autocorrelated, as were the linear model residuals, a Geographically Weighted Regression was performed. This reduced the spatial autocorrelation (Moran's I = 0.328), improved the model fit (Adj R2 = 80.5; AIC = 1857), and illustrated that the associations between socioeconomic and demographic factors and drug misuse deaths are spatially variable between LA across England. These variations are plotted in Section 4.3.2.4 and demonstrate how in some LA, a one-unit increase in the proportion of residents of white ethnicity causes drug misuse deaths to increase by 0.504 whilst in others, this impact is reduced to 0.088. Similar variability is also seen in the other variables. Such differences across LA in England suggest that there are other, as yet unaccounted for, factors that are driving drug misuse deaths. 

**5.4.** _"**What are the temporal trends in drug misuse deaths?" (RQ.4)**_

To answer RQ 4, single and double exponential smoothing techniques were employed. Despite using the optimal smoothing level, the single exponential average had a poor model fit. It exhibited much of the variation in the raw data and fails to provide much insight. This is because of the high alpha parameter used (0.995), which meant that historical observations were not heavily weighted. It is also because single exponential smoothing does not capture variation in the trend component (i.e., the steep gradients) present within the drug misuse death data. In comparison, double exponential smoothing using the above alpha parameter and an optimal beta parameter of 0.2 performed well. It demonstrates a smooth average over the time period and predicts an increase of 10% in drug misuse deaths by 2024. 

**5.5.** **Limitations, policy proposals, and future work**

The project has a number of limitations: first, some datasets are not directly comparable to 2021. These include the prevalence of common mental health disorders in an LA (2017), the proportion of households in fuel poverty (2020), and the income deprivation score (2019). These were used because it was thought that they would provide insight regardless, and because these were the only available data. Second, data for the prevalence of drug misuse contains Welsh information in addition to the English, making it less comparable.

The results of this project highlight the need for greater and more targeted drug harm reduction strategies, implemented both at a national level and within individual LA. Predicted increases in the number of drug misuse deaths means that the disparities identified between genders and age bands will likely continue to worsen. This not only has individual impacts, but entrenches the cycles of drug misuse, deprivation, and poverty that blight communities. The findings also demonstrate that harm reduction strategies are failing: current policy focuses on hard-line measures and tough judicial sentences for drug users, but the findings of this project suggest that as much as 80.5% of the variation in drug misuse deaths is accounted for by the six demographic and socioeconomic variables discussed above. Thus, perhaps, intervention and strategies aimed at reducing drug misuse deaths should aim to reduce the systematic inequalities of these variables. 

Building on these findings, future work would assess the associations between other socioeconomic and demographic variables, such as homelessness, country of birth, and occupations, and drug misuse deaths. It would also breakdown existing variables further, including by assessing different educational attainment types, rather than just degree holders. This would aid in providing a comprehensive understanding of the factors behind drug misuse deaths. In this project, we could only analyse drug misuse deaths as a function of age, gender, and drug misuse rates at a national scale due to data availability. Future work would seek to source data that allows an analysis of these variables at an LA scale. This would again complement the comprehensiveness of the findings. A further interesting pathway would be to analyse how the use of and access to specific drugs in LA affects drug misuse deaths.
